```python
def scrape_f21(df):
    """ обробка файлів:
    "роды в отд.возраст.группах" 
    з 2015 по 2020 рік
    
    """
    cols = df.iloc[4:6,:].T.iloc[2:,:].fillna(method='ffill')[4]+'|'+df.iloc[4:6,:].T.iloc[2:,:].fillna(method='ffill')[5]
    cols2 = [i for i in cols]
    col3 = ['регіон']
    col3.extend(cols2)
    res = df.iloc[8:,1:]
    res.columns = col3
    res['рік'] = df.iloc[2,2]
    return res
```

```python
df = pd.DataFrame()
for i in range(len(files)):
    df = pd.concat([df,scrape_f21(pd.read_excel(path1+files[i]))])
```
